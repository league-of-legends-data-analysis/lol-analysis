```python
import pandas as pd
```

### 최종 모델 데이터 가져옴 

```python
df = pd.read_csv('data/모델돌릴꺼2.csv')
df.head(2)
```

### Unnamed: 0 제거

```python
df.drop('Unnamed: 0',1,inplace=True)
df.head(2)
```

```python
df['라인'].unique()
```

### 각 라인별로 데이터 나누기

```python
top = df.loc[df['라인']=='TOP']
jungle = df.loc[df['라인']=='JUNGLE']
mid = df.loc[df['라인']=='MIDDLE']
bottom = df.loc[df['라인']=='BOTTOM']
utility = df.loc[df['라인']=='UTILITY']
```

### 라인 구분할때 사용했던 컬럼 제거

```python
top.drop(['라인'],1, inplace=True)
jungle.drop(['라인'],1, inplace=True)
mid.drop(['라인'],1, inplace=True)
bottom.drop(['라인'],1, inplace=True)
utility.drop(['라인'],1, inplace=True)
```
