```python
import pandas as pd
```

```python
df = pd.read_csv('data/2022_09_10000_matches_한글버전_컬럼전처리.csv')
```

```python
df.columns
```

```python
df['게임 시간 (분)'] = round(df['게임 시간 (초)']/60,2)
```

```python
df['분당 얻은 골드'] = round(df['얻은 골드'] / df['게임 시간 (분)'],2)
df['분당 피해량'] = round(df['챔피언에게 가한 피해량'] / df['게임 시간 (분)'],2)
df['분당 cc건시간'] = round(df['다른 챔피언에게 cc기를 건 시간'] / df['게임 시간 (분)'],2)
df['분당 CS'] = round(df['cs 수'] / df['게임 시간 (분)'],0)
df['분당 죽은시간'] = round(df['부활 대기 죽은 시간'] / df['게임 시간 (분)'],2)
df['분당 시야점수'] = round(df['시야점수'] / df['게임 시간 (분)'],2)
df['분당 킬'] = round(df['킬수'] / df['게임 시간 (분)'],2)
df['분당 어시'] = round(df['어시'] / df['게임 시간 (분)'],2)
df['분당 데스'] = round(df['데스'] / df['게임 시간 (분)'],2)
```

```python
df
```

```python
df = df.loc[:,['라인','win','분당 킬','분당 어시','분당 데스','퍼블유무','퍼블어시유무','분당 얻은 골드','분당 피해량','분당 cc건시간','분당 CS','분당 죽은시간','분당 시야점수']]
```

```python
df = df[df['라인'] != 'Invalid']
```

```python
df.to_csv('data/모델돌릴꺼2.csv')
```
