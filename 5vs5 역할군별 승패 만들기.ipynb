```python
#일단 팀구분하고 1만개 경기라고 생각함.
duo_stats1 = pd.DataFrame()
for i in range(0,len(df),10):
    stat={}
    stat['GameID'] = df.loc[i,'GameID']
    stat['Win'] = df.loc[i,'win']
    for j in range(12):
        stat[check_bluestr[j]] = 0
        stat[check_redstr[j]] = 0
    
    for j in range(0,5):
        k = df.loc[i+j,'챔피언']
        for check_idx in range(12):
            if k in check[check_idx]:
                stat[check_bluestr[check_idx]]+=1
    for j in range(5,10):
        k = df.loc[i+j,'챔피언']
        for check_idx in range(12):
            if k in check[check_idx]:
                stat[check_redstr[check_idx]]+=1
    duo_stats1 = pd.concat([duo_stats1, pd.Series(stat)], axis=1, sort=False)
duo_stats1 = duo_stats1.T.reset_index(drop=True)
    
```
